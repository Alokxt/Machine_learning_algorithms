```python
from sklearn.preprocessing import StandardScaler
```

```python
def PCA(data_frame,thresh=None):
    mean = np.mean(data_frame,axis=0)
    
    Xc = StandardScaler().fit_transform(data_frame)

    Cov = np.cov(Xc,rowvar=False)
    Eig_vals,Eig_vecs = np.linalg.eig(Cov)
    indexes = np.argsort(Eig_vals)[::-1]
    sorted_eig_val = Eig_vals[indexes]
    sorted_eig_vec = Eig_vecs[:,indexes]
    print(sorted_eig_val)
    if thresh is None:
        thresh = 0.92
    total = np.sum(sorted_eig_val)
    cum_var = np.cumsum(sorted_eig_val) / total
    top_k = np.argmax(cum_var >= thresh) + 1

    print(top_k)
    final_eigs_vecs = sorted_eig_vec[:,:top_k]

    cols = [f'{i+1}th PC' for i in range(top_k)]
    
    Projected_data = np.dot(Xc,final_eigs_vecs)
    new_Data_frame = pd.DataFrame(Projected_data,columns=cols,index = data_frame.index)
    return new_Data_frame
    
```
