```python
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
from sklearn.datasets import load_diabetes
data = load_diabetes()
X,Y = data.data , data.target
```

```python
def gradeint(x,y,w):
    m = x.shape[0]
    return (x.T@x)@w - (x.T@y)
```

```python
def gradient_descent(x,y,n,alpha):
    
    
    wi = np.zeros(x.shape[1],dtype=float)
    for i in range(n):
        wi -=  alpha*(gradeint(x,y,wi))
    
    return wi 
```

```python
def predict_points(x,w,intercept):
    y_pred = np.zeros(x.shape[0],dtype=float)
    for i in range(len(x)):
        pred = np.dot(x[i],w)
        y_pred[i] = pred +intercept 
    return y_pred 
    
```

```python
X_train,X_test = X[:340,], X[340:,]
```

```python
Y_train,Y_test = Y[:340],Y[340:]
```

```python
W = gradient_descent(X_train,Y_train,50,1e-10)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python

x_scale = X_train- np.mean(X_train,axis=0)
```

```python
W2 = gradient_descent(x_scale,Y_train,50,1e-1)
```

```python
inter = np.mean(Y_train) - np.mean(X_train,axis=0)@W2
```

```python
y_preds2 = predict_points(X_train,W2,inter)
```

```python
print(mean_squared_error(Y_train,y_preds2))
```

```python
r2_score(Y_train,y_preds2)
```

```python
y_preds_test = predict_points(X_test,W2,inter)
```

```python
print(mean_squared_error(Y_test,y_preds_test))
```

```python
r2_score(Y_test,y_preds_test)
```
